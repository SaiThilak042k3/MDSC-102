```python
def Plotting(features):    
    
    plt.figure(figsize=(15,5))

    #BOX PLOT
    plt.subplot(1,3,1)
    plt.xlabel('X Label')
    sns.boxplot(data= df, x= features, color='red').set(title="Box Plot for "+features)
    
    #HISTPLOT
    plt.subplot(1,3,2)
    sns.histplot(df[features], kde=True, color='blue').set(title="Histogram for " + features)   
    
    #PROBABILITY PLOT
    plt.subplot(1,3,3)
    plt.title('Probability Plot for' + features)
    sp.stats.probplot(df[features], plot=plt)
```

```python
#beforing Normaling the Features

for j in df.columns:
    if(j != 'Class'):
        Plotting(j)
```

# **Inference Skewness**





> 

*    Expect Malic Acid and Magnesium all other features are near to zero.
*   This shows that all other points are almost normalizied.
*   By using the **BoxCox Function** will find the lambda value and using formula  will normalize the features


```python
lambdaList={}

for i in df.columns:
    if(i != 'Class'):
        #finding lamda values
        feat , a = stats.boxcox(df[i])
        lambdaList[i] = a
        value = ((df[i]**a) -1) / a
        df[i] = value

        #after Normalizing
        Plotting(i)

```

# **Inference After Normalizing**



>    
*   After normalizing the features all the plots clearly shows that it is normalizied (almost) and the shewness also reduced (leading to zero).
