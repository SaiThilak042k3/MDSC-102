```python
#importing libraries

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.stats import binom
```

 ## Binomial 



```python
# Define the parameters
n= int(input("Enter the N = "))  # Number of trials
p= float(input("Enter the P = "))  # Probability of success


X_values = np.arange(0,n+1)  #possible vaules x can assume


probabilities = np.array([np.math.comb(n, x) * (p**x) * ((1-p)**(n-x)) for x in X_values])


plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution")
#plt.plot(X_values,probabilities)
plt.bar(X_values,probabilities)





```

 ## Possion 
 

```python
# Define the parameters
lamda= float(input("Enter the N = "))  # lamda

X_values = np.arange(0,n+1)  #possible vaules x can assume


probabilities = np.array([(np.exp(-lamda) * lamda**x) / np.math.factorial(int(x)) for x in X_values])

plt.xlabel("X")
plt.ylabel("Probability")
plt.title("Possion Distribution")
#plt.plot(X_values,probabilities)
plt.bar(X_values,probabilities)

```

## Exponential 

```python
# Define the parameters
lamda= float(input("Enter the N = "))  # lamda

X_values = np.arange(-10,11,.2)  #possible vaules x can assume

probabilities = np.array([(lamda * np.exp(-lamda*x)) for x in X_values])

plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution")
#plt.plot(X_values,probabilities)
plt.plot(X_values,probabilities)
```

## Normal

```python
# Define the parameters
u= float(input("Enter the standard deviation = ")) 
o= float(input("Enter the mean = "))  

pi = 3.1415

X_values = np.arange(-10,11,.2)  #possible vaules x can assume



probabilities = np.array([1/np.sqrt(2*pi*o**2)*np.exp(-(x-u)**2/2*o**2) for x in X_values])


plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.plot(X_values,probabilities)
#plt.bar(X_values,probabilities)
```

## Gamma

```python
# Define the parameters
k= float(input("Enter the alpha = ")) 
theta= float(input("Enter the lamda = "))


X_values = np.linspace(0, 10, 1000) #possible vaules x can assume

def gamma_pdf(x, k, theta):
    numerator = (x ** (k - 1)) * np.exp(-x * theta)
    denominator = (theta ** k) 
    return numerator / denominator


# Calculate the manual gamma PDF for each x value
probabilities = np.array([gamma_pdf(x,k, theta) for x in X_values])


#plt.plot(x, pdf, label=f"Gamma(k={k}, theta={theta})")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Gamma Distribution")
plt.plot(X_values,probabilities)
#plt.bar(X_values,probabilities)
plt.show()
```
