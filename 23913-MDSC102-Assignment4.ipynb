# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

```python

sns.countplot(x=df['Different Wines']).set_title('Countplot Example') # Count plot example
```

The above plotting method is know Count Plot which take one variable as argument and display the count of the that particular variable

# scatter plot
x, y = vectors or keys in data
Variables that specify positions on the x and y axes.

hue = vector or key in data
Grouping variable that will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.

size = vector or key in data
Grouping variable that will produce points with different sizes. Can be either categorical or numeric, although size mapping will behave differently in latter case.

```python
sns.set_style('darkgrid')
sns.scatterplot(x=df['Malic acid'],y=df['Alcohol'],hue=df['Different Wines'],size=df['Different Wines'])  # Example for scatter plot from seaborn 
```

**Inference**

**Y= Alcohol**

**X= Malic Acid**


> 

*   *Class - 1* =  less Malic acid leads to more alcohol.
*   *Class - 2* =  less Malic acid leads to less alcohol.
*   *Class - 3* =  saddle points No inference can be made.

```python
# A fucntion defined to recursively call the functions for Alcohol as Y and other different products as X
def functionScatterPlot(x,i,y='Alcohol'): 
    plt.subplot(4,3,i)  #subplot to produces 4 rows and 3 colunms plots
    sns.scatterplot(x=df[x],y=df[y],hue=df['Different Wines'],size=df['Different Wines'])
    
```

```python
i=1
#figure, axes = plt.subplots(4, 3, sharex=True, figsize=(26, 10))
plt.figure(figsize=(20, 20))
for col in df.head():
      
    if(col !='Alcohol'):  # expect Alcohol and Different Wines all other products are selected as X values 
        if( col != 'Different Wines'):
            functionScatterPlot(col,i)
            i+=1
```
