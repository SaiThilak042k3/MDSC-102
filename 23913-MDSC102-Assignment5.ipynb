# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

## Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
# for two variables
sns.boxplot(data=df, x="Different Wines", y="Alcohol")
```

```python
sns.boxplot(x=df["Alcohol"]) # for one variables
```

**Box Plots**

```python
#user defined function to generate 13 different plots
def funBoxPlot(y,i,x='Different Wines'): 
    plt.subplot(7,2,i)  #subplot to produces 4 rows and 3 colunms plots
    sns.boxplot(data=df, x=x, y=y)

```

```python
# function

i=1
plt.figure(figsize=(20, 40))
for col in df.head():      
    # expect  Different Wines all other items are selected as Y values 
    if( col != 'Different Wines'):
        funBoxPlot(col,i)
        i+=1
```

# Histplot 

**We ploting the histogram with the box the blox plot for better understanding**

```python
i=1
plt.figure(figsize=(30, 80))
for col in df.head():      
    # expect  Different Wines all other items are selected as Y values 
    if( col != 'Different Wines'):
        for y in range(1,4):
            plt.subplot(13,4,i)
            sns.histplot(x=df[df['Different Wines']==y][col],hue=df['Different Wines'])
            i+=1
        plt.subplot(13,4,i)
        sns.boxplot(data=df, x='Different Wines', y=col)
        i+=1
```
